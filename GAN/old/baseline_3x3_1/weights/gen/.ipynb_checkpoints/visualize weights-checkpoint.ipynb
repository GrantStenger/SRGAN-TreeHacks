```python
import tensorflow as tf
from keras.models import load_model
import keras.backend as K
```

```python
def root_mean_squared_error(y_true, y_pred):
    """ Defines the RMSE loss function.

        Args:
            y_true: The true label.
            y_pred: The predicted label.

        Returns:
            error: The RMSE error value

    """

    error = K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1))
    return error
```

```python
model3 = load_model("epoch_3.h5py", custom_objects={"tf":tf, "root_mean_squared_error":root_mean_squared_error})
model2 = load_model("epoch_2.h5py", custom_objects={"tf":tf, "root_mean_squared_error":root_mean_squared_error})
```

```python
weights3 = model3.get_weights()
weights2 = model2.get_weights()
```

```python
from keras.layers import  Conv2D, UpSampling2D, Dropout, Input, Lambda, Concatenate
import keras.backend as K
from keras.activations import softplus
from keras.models import Model


def cnn_transform(inval, n_filters, kernel_size=(2,2)):
    
    transformed = Conv2D(n_filters, kernel_size=kernel_size, padding="SAME", activation=softplus )(inval)
    summed = Lambda(lambda x: K.expand_dims( K.sum(x, axis=-1), ) ) (transformed)
    
    return summed


img = Input(shape=(240,426,3))

upsample = UpSampling2D()(img)

GBR = []
for i in range(3):
    
    out = cnn_transform(upsample, n_filters=8)
    GBR.append(out)

out_img = Concatenate()(GBR)
    

model = Model(inputs=[img], outputs=[out_img])

```
